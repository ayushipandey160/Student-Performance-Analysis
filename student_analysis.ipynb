# Student Performance Dataset Analysis

Maincrafts Technology Internship Task 1

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```
df = pd.read_csv('student-mat.csv', sep=';')
df.head()
```

## Explore and Clean Data

```
print('Shape:', df.shape)
print(df.info())
print(df.isnull().sum())
print('Duplicate Records:', df.duplicated().sum())
df = df.drop_duplicates()
```

## Analysis Questions

```
avg_grade = df['G3'].mean()
print('Average Final Grade (G3):', round(avg_grade,2))

above_15 = len(df[df['G3'] > 15])
print('Students scoring above 15:', above_15)

corr = df['studytime'].corr(df['G3'])
print('Correlation between Study Time and G3:', round(corr,3))

gender_avg = df.groupby('sex')['G3'].mean()
print('\nAverage Grade by Gender')
print(gender_avg)
```

## Visualizations

```
plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10, edgecolor='black')
plt.title('Distribution of Final Grades')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.show()
```

```
plt.figure(figsize=(8,5))
plt.scatter(df['studytime'], df['G3'])
plt.title('Study Time vs Final Grade')
plt.xlabel('Study Time')
plt.ylabel('G3')
plt.show()
```

```
gender_avg.plot(kind='bar')
plt.title('Average Final Grade by Gender')
plt.xlabel('Gender')
plt.ylabel('Average G3')
plt.show()
```

## Conclusion

- Dataset loaded successfully.
- Missing values and duplicates checked.
- Average final grade calculated.
- Students scoring above 15 identified.
- Correlation between study time and grades analyzed.
- Gender-wise performance compared.
- Visualizations created for better understanding.
